```python
x = news[["daily"]] # always x should be in dataframe format mandatory
y = news["sunday"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=12)
#print(x_train,x_test,y_train,y_test)
x_train = pd.DataFrame(x_train)
print(x_train)
x_test = pd.DataFrame(x_test)
print(x_test)
```

```python
from sklearn.linear_model import LinearRegression # for both simple and multi linear model is linear regreassion
```

```python
slr = LinearRegression() # building the model usin linear regression
slr.fit(x_train,y_train) # Training the model with question and answer
```

```python
# model testing
y_train_pred = slr.predict(x_train)
y_test_pred = slr.predict(x_test)
```

```python
x_train["y_train"] = y_train
x_train["y_train_pred"] = y_train_pred
x_train
```
